```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import keras
from keras.datasets import cifar100
from keras.preprocessing.image import ImageDataGenerator
from cutout_generator import CutoutImageDataGenerator
from mixup_generator import MixupGenerator
from cutmix_generator import CutMixGenerator
import tensorflow as tf
```

```python
(x_train, y_train), (x_test, y_test) = cifar100.load_data()
num_classes = 100
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
datagen0=ImageDataGenerator()
generator0 =datagen0.flow(x_train, y_train,shuffle=False,batch_size=3)
generator1 = MixupGenerator(x_train, y_train, shuffle=False,alpha=0.4,batch_size=3)()
datagen = CutoutImageDataGenerator(cutout_mask_size=16)
generator2 =datagen.flow(x_train, y_train,shuffle=False,batch_size=3)
generator3 = CutMixGenerator(x_train, y_train,shuffle=False, alpha=0.8,batch_size=3)()
```

```python
cols, rows = 3, 1
for x,y in generator0:
    i=0
    for i in range(3):
        #plt.imsave(str(i)+"_0.png",x[i].astype(np.uint8))
        plt.subplot(rows, cols, i + 1)
        plt.imshow(x[i].astype(np.uint8), interpolation="nearest")
        plt.axis('off')
    break
```

```python
cols, rows = 3, 1
for x,y in generator1:
    i=0
    for i in range(3):
        #plt.imsave(str(i)+"_1.png",x[i].astype(np.uint8))
        plt.subplot(rows, cols, i + 1)
        plt.imshow(x[i].astype(np.uint8), interpolation="nearest")
        plt.axis('off')
    break
```
